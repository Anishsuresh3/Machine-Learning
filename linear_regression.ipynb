```python
def compute(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b;
    return f_wb
```

```python
#manually setting the parameters w and b
t_wb = compute(x_train,9116.1271,26577.3467)
t_wb
```

```python
fig,ax=plt.subplots()
ax.plot(x_train,t_wb)
scatter = ax.scatter(x_train,y_train,marker='X')
plt.show()
```

```python
def compute_cost(x,y,w,b):
    m = x.shape[0]
    cost_sum=0
    for i in range(m):
        tmp_f = w*x[i]+b
        cost=(tmp_f-y[i])**2
        cost_sum+=cost
    total_cost = cost_sum/(2*m)
    return total_cost
```

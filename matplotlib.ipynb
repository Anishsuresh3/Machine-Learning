## Making figures with NumPy arrays

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
#returns evenly spaced numbers over a specified interval
x[:10]
```

```python
# Plot the data and create a line plot
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# Scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Making a plot from dict
cars = {"nissan":20,
        "BMW":26,
        "audi":30}
fig,ax=plt.subplots()
ax.bar(cars.keys(),cars.values())
ax.set(title="Car Sales",
       ylabel="No. of cars sold");
```

```python
fig,ax=plt.subplots()
ax.barh(list(cars.keys()),list(cars.values()));
```

```python
# Makeing histogram
x=np.random.randn(1000) #Return a sample (or samples) from the "standard normal" distribution.
fig,ax=plt.subplots()
ax.hist(x);
```

## Two Options for subplots

```python
#subplot option 1
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(cars.keys(),cars.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(cars.keys(),cars.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

* When plotting something quickly, can use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_40 = heart_disease[heart_disease["age"]>40]
over_40.head()
# len(over_40) 284
```

```python
#pyplot method
over_40.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
# OO method
fix, ax=plt.subplots(figsize=(10,6))
over_40.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
# ax.set_xlim([35,100])
```

```python
# OO method frm scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_40["age"],y=over_40["chol"],c=over_40["target"]);
#customize
ax.set(title="Heart Disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Cholestrol");
#add legend
ax.legend(*scatter.legend_elements(),title="Target")
#add a horizontal line
ax.axhline(over_40["chol"].mean(), linestyle='--') #Add a horizontal line across the Axes.
#Parameters
# y : float, default: 0
    # y position in data coordinates of the horizontal line.
```

```python
heart_disease.head()
```

```python
fig, ax=plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_40["thalach"],y=over_40["chol"],c=over_40["target"])
#customizing
ax.set(title="Max heart rate and Cholestrol levels",
       xlabel="Max Heart rate",
       ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_40["chol"].mean(),linestyle='--');
```

```python
# subplot of chol,thalach,age
fig, (ax0 ,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter=ax0.scatter(x=over_40["age"],y=over_40["chol"],c=over_40["target"])

ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="CHolestrol")
ax0.legend(*scatter.legend_elements(),title="Target")

ax0.axhline(over_40["chol"].mean())

scatter=ax1.scatter(x=over_40["age"],y=over_40["thalach"],c=over_40["target"])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_40["thalach"].mean(),linestyle='--');
# adding a title to the fig
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
